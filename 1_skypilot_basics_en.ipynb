## Verifying Cloud Access
In the prereqs we configured our AWS account. We are now going to check if SkyPilot recognizes AWS.

```python
! sky check aws
```

You should get a message that says something like `AWS [compute, storage]`, meaning that SkyPilot *recognizes* AWS as an infra provider, and that SkyPilot has access to the necessary credentials to perform necessary operations.
